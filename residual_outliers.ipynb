```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uncleaned_wetchem_df = pd.read_csv("uncleaned_wetchem.csv")
uncleaned_wetchem_df = uncleaned_wetchem_df.rename(columns={"Unnamed: 0":"sample_code"})
uncleaned_wetchem_df.set_index("sample_code")
```

```python
for column in uncleaned_wetchem_df.columns:
    if(column != 'sample_code'):
        vals = []
        for value in uncleaned_wetchem_df[column].values:
            if(value is not None):
                value = str(value)
                value = value.replace(">","").replace("<","").replace("...","").strip()
                value = float(value)
            vals.append(value)
        uncleaned_wetchem_df[column] = vals
```

```python
wetchem_df = uncleaned_wetchem_df.copy(deep=True)
wetchem_df.set_index("sample_code")
```

```python
outliers = {'boron' : [-5, 5], 'phosphorus' : [-250, 450],'zinc' : [-25, 100], 'sulphur' : [-100, 400],'sodium':[-1000,2500], 'magnesium':[-500,1000],'potassium' : [-800,1600],'calcium':[-5000,5000],'copper' : [-100,300],'ec_salts' : [-1000,2000],'organic_carbon' : [-2,2] }
```

```python
pcc3_elements = ['cec', 'copper', 'iron', 'manganese', 'ph', 'total_nitrogen']
```

```python
pcc_classes_dict = pd.read_csv("element_managemet_thresholds.csv").T.to_dict()
```

```python
all_chemicals = ['psi', 'aluminium', 
            'phosphorus', 'ph', 'exchangeable_acidity', 'calcium', 'magnesium',
              'sulphur', 'sodium', 'iron', 'manganese', 'boron', 'copper', 'zinc', 'total_nitrogen', 'potassium',
             'ec_salts', 'organic_carbon', 'cec', 'sand', 'silt', 'clay']
```

```python
wetchem_df = wetchem_df.reset_index()
```

```python
outlier_sample_codes = []
per_chem_outliers = {}
final_outlier_df = pd.DataFrame()
final_outlier_df['chemical'] = all_chemicals
for file in os.listdir("outputFiles/preds"):
    if(".csv" in file):
        chem = file.split('.')[0]
        print(chem)
        df = pd.read_csv(f"./outputFiles/preds/{file}")
        if('sample_code' not in df.columns):
            df = df.rename(columns={'sample_id':'sample_code'})

        df = pd.merge(df, wetchem_df, on='sample_code', how="inner")
        if(chem in outliers.keys()):
            print(f"DF Before outliers {len(df.index)}")
            outliers_df = df.loc[(df['0'] <= outliers[chem][0]) | (df['0'] >= outliers[chem][1])]
            outlier_sample_codes.append(outliers_df['sample_code'].values)
            print(f"outliers {chem}",len(outliers_df.index))
            final_outlier_df.loc[final_outlier_df['chemical'] == chem, 'redbooth_defined_outliers'] = len(outliers_df.index)
        elif(chem in pcc3_elements):
            chem_df = df[['sample_code',chem, '0']]
            df = chem_df
            lower = None
            mid = None
            upper = None
            for i in pcc_classes_dict.keys():
                if(chem in pcc_classes_dict[i]['soilProperty']):
                    lower = pcc_classes_dict[i]['Value_1']
                    mid = pcc_classes_dict[i]['Value_2']
                    upper = pcc_classes_dict[i]['Value_3']
                    break
            if(chem in df.columns):
                print("Return....",type(df.loc[df[chem] > 0]))
                df.loc[df[chem] >= upper, "actual_class"] = 4
                df.loc[df['0'] >= upper, "predicted_class"] = 4
                
                df.loc[(df[chem] >= mid) & (df[chem] < upper), "actual_class"] = 3
                df.loc[(df['0'] >= mid) & (df['0'] < upper), "predicted_class"] = 3

                df.loc[(df[chem] >= lower) & (df[chem] < mid), "actual_class"] = 2
                df.loc[(df['0'] >= lower) & (df['0'] < mid), "predicted_class"] = 2

                df.loc[df[chem] < lower, "actual_class"] = 1
                df.loc[df['0'] < lower, "predicted_class"] = 1
    #             joined_df.to_csv(f"../QC_Model Preds_2022-11-06/preds/{file}")
                df = df.loc[(df['actual_class'] - df['predicted_class']) >= 2]
                print(f"PCC3 {chem}",len(df.index))
                outlier_sample_codes.append([*df['sample_code'].values])
                final_outlier_df.loc[final_outlier_df['chemical'] == chem, 'pcc3_outliers'] = len(df.index)

            continue
        else:
            print('')

            
            
        
```
