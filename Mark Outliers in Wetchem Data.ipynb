```python
import pandas as pd
import numpy as np
```

```python
wetchem_df = pd.read_csv("all_wetchem_data_uncleeaned_15-7-2022 (1).csv")
hv_outliers_df = pd.read_csv("outputFiles/hv_outliers.csv")
lv_outliers_df = pd.read_csv("outputFiles/lv_outliers.csv")
resid_outliers_df = pd.read_csv("outputFiles/residual_outliers_filtered_out.csv")
cooks_outliers_df = pd.read_csv("outputFiles/outliers_cooks_distance.csv")
```

```python
hv_outliers_df
```

```python
lv_outliers_df
```

```python
resid_outliers_df
```

```python
cooks_outliers_df
```

```python
wetchem_df = wetchem_df.rename(columns={"Unnamed: 0":"sample_code"})
```

```python
# Handle the HV Outliers
hv_outliers_df_2 = hv_outliers_df.set_index("sample_code").isna()   
```

```python
hv_outliers_df_2.replace(True, np.nan, inplace=True)
hv_outliers_df_2.replace(False, 1, inplace=True)
```

```python
hv_outliers_df_2
```

```python
# Handle lv outliers
lv_outliers_df_2 = lv_outliers_df.set_index("sample_code").isna()   

lv_outliers_df_2.replace(True, np.nan, inplace=True)
lv_outliers_df_2.replace(False, 2, inplace=True)
```

```python
lv_outliers_df_2
```

```python
# Handle Cook's outliers
cooks_outliers_df_2 = cooks_outliers_df.set_index("sample_code").isna()   

cooks_outliers_df_2.replace(True, np.nan, inplace=True)
cooks_outliers_df_2.replace(False, 3, inplace=True)
```

```python
cooks_outliers_df_2
```

```python
# Handle Residual outliers
resid_outliers_df_2 = resid_outliers_df.set_index("sample_code").isna()   

resid_outliers_df_2.replace(True, np.nan, inplace=True)
resid_outliers_df_2.replace(False, 4, inplace=True)
```

```python
resid_outliers_df_2
```

```python
wetchem_df
```

```python
wetchem_df = wetchem_df[['sample_code']]
```

```python
wetchem_df
```

```python
outliers_df = pd.concat([hv_outliers_df_2, lv_outliers_df_2, cooks_outliers_df_2, resid_outliers_df_2], axis=0)
```

```python
outliers_df = outliers_df.reset_index()
outliers_df
```

```python
outliers_df = outliers_df.groupby("sample_code", sort=True).min()
```

```python
outliers_df
```

```python
outliers_df = pd.merge(wetchem_df,outliers_df, on="sample_code", how="left")
```

```python
outliers_df
```

```python
outliers_df.to_excel("outputFiles/dl_outliers.xlsx")
```
